```python
import matplotlib.pyplot as plt
import math  as m

#initial values
duration = 5000
nx = 100
delta_x = 2  / nx
d =  0.1

x =  []
y = []

#setup initial distribution
for i in range(nx):
    x.append(-1 + i * delta_x)
    y.append(m.exp(-((x[i]**2)/d)))

dataX = []
dataY = []
dataZ = []

def central(i):
    return (y[(i-1) % nx] - y[(i+1) % nx]) / 2 * delta_x

for i in range(duration):
    for j in range(nx):
        y[j] += central(j)

        if i % 1000 == 0:
            dataX.append(x[j])
            dataY.append(y[j])
            dataZ.append(i)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.set_xlabel('x', fontweight='bold')
ax.set_ylabel('t', fontweight='bold')
ax.set_zlabel('f(x)', fontweight='bold')

ax.scatter(dataX, dataZ, dataY)
plt.show()
```
